```python
import numpy as np    #serve para manipulacao numerica em python
import pandas as pd # Import pandas here

#criacao de dataframe

df_salarios = pd.DataFrame({
    'nome':["livia", "joa", "maria", "Rui"],
    'salario':[4000, np.nan , 3000, np.nan]
})
#filna serve para preencher valores nulos
#calcula media salarial e substitui os nulos pela media e arrendomda os valores
#mean- media
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))
#calcula mediana
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].medean().round())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    "Dia": ["Segunda", "Terca", "Quarta", "Quinta", "sexta"],
    "Temperatura": [30, np.nan, np.nan, 28, 27]
})
#forward ffill preenche os espa;os vazios com o calor anterior bfill prenche com o valor posterior

df_temperaturas["preenchido_ffill"] = df_temperaturas["Temperatura"].ffill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nome':["livia", "joa", "maria", "Rui", "liz"],
    'cidade': ["Maputo", np.nan, "Nampula", np.nan, "vale"]
})

df_cidades['cidade_preenchida'] = df_cidades["cidade"].fillna("nao informado")
df_cidades
```
